```python
# create map of TORONTO using latitude and longitude values above:
map_oc = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(result_high['location.lat'], result_high['location.lng'], result_high['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_oc)  

for lat, lng, label in zip(result_low['location.lat'], result_low['location.lng'], result_low['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color='#000000',
        fill_opacity=0.7,
        parse_html=False).add_to(map_oc)      
    
map_oc
```

```python

from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
##result=result.drop("Cluster Labels",axis=1)
```

```python
k=5
oc_clustering = result[['location.lat','location.lng']]
kmeans = KMeans(n_clusters = k,random_state=0).fit(oc_clustering)
kmeans.labels_
result.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
map_clusters = folium.Map(location=[33.686100,-117.826280],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(result['location.lat'], result['location.lng'], result['name'], result['Cluster Labels']):
    label = folium.Popup(' Name ' + str(neighbourhood), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
    
for lat, lng, label in zip(result_low['location.lat'], result_low['location.lng'], result_low['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=9,
        popup=label,
        color='red',
        fill=True,
        fill_color='#000000',
        fill_opacity=0.0,
        parse_html=False).add_to(map_clusters)
    
for lat, lng, label in zip(result_high['location.lat'], result_high['location.lng'], result_high['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=9,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.0,
        parse_html=False).add_to(map_clusters)
       
map_clusters
```
